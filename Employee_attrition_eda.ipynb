```python
# Setup Cell (run first)
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style explicitly
sns.set(style="whitegrid")
```

```python
import pandas as pd
df=pd.read_csv("E:/Data science/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
print(df.describe(include="all"))
```

```python
print("Shape",df.shape)
```

```python
print("Datatypes",df.dtypes)
```

```python
categorical= df.select_dtypes(include='object').columns.tolist()
numerical= df.select_dtypes(include='int64').columns.tolist()
print("categorical",categorical)
print("numerical",numerical)
```

```python
print("missing values", df.isnull().sum())
```

```python
for col in df.columns:
    if df[col].nunique==1:
        print(f"dropping the constant column:{col}")
        df.drop(col,axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 4))
sns.boxplot(x=df["MonthlyIncome"])
plt.title("Boxplot of Monthly Income")
plt.show()
```

```python
df[numerical].hist(figsize=(10,12),bins=30)
plt.suptitle("Histogram for Numeric Columns",fontsize=16)
plt.tight_layout()
plt.show()
```

```python
for col in categorical:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df)
    plt.title(f"Countplot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
print(df["Attrition"].value_counts())
sns.countplot(x='Attrition', data=df)
plt.title("Attrition Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[numerical].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Monthly Income by Attrition")
plt.show()
```

```python
 plt.figure(figsize=(10, 5))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title("Job Role vs Attrition")
plt.xticks(rotation=45)
plt.show()
```

```python
grouped = df.groupby('Department')['YearsAtCompany'].mean().sort_values()
grouped.plot(kind='bar', title="Average Years at Company by Department", figsize=(8, 5))
plt.ylabel("Average Years")
plt.show()
```
